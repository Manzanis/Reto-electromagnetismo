# Gráfica

```python
def graph (x1,x2,f):
    x=np.linspace(x1,x2,50)
    fig=plt.figure(figsize=(12,4))
    fig = plt.plot(x,f(x))
    plt.plot(x,np.zeros((50,)))
```

# Bisección

```python
def bisection (f,a,b):
    d=1
    fa=f(a)
    fb=f(b)
    for i in range(500):
        fa=f(a)
        fb=f(b)
        c=(a+b)/2
        fc=f(c)
        if ((fa*fc)<0):
            b=c
        if ((fb*fc)<0):
            a=c
        if (abs(fa-fb)<2e-12):
            return c
    
```

# Newton

```python
def diff (f,x,deltax=1e-12):
    return((f(x+deltax)-f(x)))/deltax
```

```python
def newnewton (f,x):
    for i in range (500):
        
        x_1=x-((f(x))/(diff(f,x)))
        
        if(abs(x-x_1)<1e-12):
            break
        
        x=x_1
    return x_1
```
